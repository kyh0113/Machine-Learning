데이터셋 가져오기

```python
import pandas as pd

data=pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data',header=None)
data.columns=['sepal length', 'sepal width', 'petal length', 'petal width', 'class']

data.head()
```

공분산과 상관계수를 통해 두 변수가 서로 관련이 있는지, 또는 독립적인지에 대한 연관성을 판단할 수 있다.

양의 연관성 :  x가 커지면 y도 커진다.

ex ) 꼭 그런건 아니지만 평균적으로 키(x)가 크면 몸무게(y)도 많이 나간다.

여러 사람의 데이터를 산점도로 나타낸 후 이 관계를 직선으로(일차함수)로 나타냈을때 기울기가 양수면 양의 상관이라 한다.

음의 연관성 :  x가 커지면 y는 작아진다.

ex) 담배(x)를 많이 피우면 수명(y)이 짧아진다.

여러 사람의 데이터를 산점도로 나타낸 후 이 관계를 직선으로(일차함수)로 나타냈을때 기울기가 d음수면 음의 상관이라 한다.

무관성 : 두 관계가 서로 관련이 없는 독립적이다.

ex) 머리(x)가 좋으면 밥을 잘 먹는다(y).

기울기가 0에 가깝다

x의 분산: x에서 x의 모평균을 빼준 후 그걸 제곱한 값에 기댓값 : x가 얼마나 퍼져있는가

y의 분산: y에서 y의 모평균을 빼준 후 그걸 제곱한 값의 기댓값 : y값들이 얼마나 퍼져이는가

x,y의 공분산은 x에서x평균 빼준거랑 y에서 y평균 빼준거 곲한 값의 기대값 : x,y가 어떠한 방향성을 가졌는지 표현한것

관계가 더 밀접해도 범위가 작으면 공분산값이 적게 나옴
단위,범위에 영향을 받기때문에 표준화 시켜줄 필요가 있음
표준화해줄 궁극적인 연관성의 측도가 상관계수이다

상관계수는 -1<r<1

r이 0에 가까우면 두 변수가 별로 관련성이 없다.

r값이 1에 가까우면 양의 상관관계가 높고 -1에 가까우면 음의 상관관계가 높다

각 양적 속성에 대해 평균, 표준 편차, 최소값 및 최대값을 계산하기

```python
from pandas.api.types import is_numeric_dtype

for col in data.columns:
  if is_numeric_dtype(data[col]):
    print('%s:' % (col))
    print('\t Mean = %2.f' % data[col].mean())
    print('\t Standard deviation = %2.f' % data[col].std())
    print('\t Minimum = %.2f' % data[col].min())
    print('\t Maximum = %2.f' % data[col].max())
```

corr()함수는 모든 변수 쌍 사이의 상관계수를 계산한다.

표의 대각선을 보면 전부 1이라는건 두 변수가 매우 관련성이 있다는 뜻 특히 양의 상관성이 높다

정성 속성(클래스)의 경우 각 고유 값에 대한 빈도를 계산하기

```python
data['class'].value_counts()
```

```python
data.describe(include='all')
```

다변량 통계량의 경우 속성 쌍 간의 공분산 및 상관 관계를 계산할 수 있다.

```python
print('Covariance:')
data.cov()
```

```python
print('Correlation:')
data_corr=data.corr()
print(data_corr)
```

양의 상관관계

```python
data_corr.loc[data_corr['petal length']>0, 'petal length'] # petal length와 관련된 양의 상관관계 출력
```

음의 상관관계

```python
data_corr.loc[data_corr['petal length']<0, 'petal length']
```
